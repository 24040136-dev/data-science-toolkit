```python
import numpy as np
arr_2x2 = np.array([[1, 2], [3, 4]])
print("2x2 array:\n", arr_2x2)
```

```python
arr_random_3x4 = np.random.rand(3, 4)
print("3x4 random array:\n", arr_random_3x4)
```

```python
arr_ones = np.ones((2, 3))
arr_zeros = np.zeros((4, 2))
print("Array of ones:\n", arr_ones)
print("Array of zeros:\n", arr_zeros)
```

```python
arr_6x3 = np.random.rand(6, 3)
arr_transposed = arr_6x3.T  # or arr_6x3.transpose()
print("Original 6x3 array:\n", arr_6x3)
print("Transposed array:\n", arr_transposed)
```

```python
n, p, m = 2, 3, 4
arr1 = np.random.rand(n, p)
arr2 = np.random.rand(p, m)
dot_product = np.dot(arr1, arr2)  # or arr1 @ arr2
print("Array 1 (2x3):\n", arr1)
print("Array 2 (3x4):\n", arr2)
print("Dot product result (2x4):\n", dot_product)
```
